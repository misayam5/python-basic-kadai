```python
class Human:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name + "は大人です")
        else:
            print(self.name + "は大人ではありません")


humans = [
    Human("太郎", 25),
    Human("花子", 18),
    Human("次郎", 30)
]

for human in humans:
    human.check_adult()
```
